<h2> <u> Heat Maps </u> </h2>

In this demo we will perform the following operations:

- Title:

    - Set the title as"Heatmap of Correlatio", set the fontsize as 20
    
- x-axis operations:
    - Set x-tick marks equal to the dimension of the columns of the correlation matrix
    - Set the x-tick marks fontsize to 10, rotation to 90

- y-axis operations:
    - Set y-tick marks equal to the dimension of the rows of the correlation matrix
    - set the y-axis fontsize to 10

- Eliminate the top and right hand spine

- Annotate the individual pixel block with correlation value

- Display the colormap

<h6> Importing Libraries </h6>

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

<h6> Reading Datasets </h6>

```python
df = pd.read_csv("housing.csv")
```

<h6> Dataset operations and Correlation operation </h6>

```python
df.drop(["longitude", "latitude", "ocean_proximity"], axis=1, inplace=True)
df_corr = np.array(df.corr())
```

<h6> Plotting the Graph </h6>

```python
axis_labels = ["median-age", "total-rooms", "Total-Bedrooms", "Population", "No of house holds", "median-income", "median-house-value"]

fig,ax = plt.subplots()

im = ax.imshow(df_corr)

plt.title("Heatmap of Correlation", fontsize=20) # Title

# X-axis Operations
ax.set_xticks(np.arange(7))
ax.set_xticklabels(axis_labels, fontsize=10, rotation=90)

# Y-axis Operations
ax.set_yticks(np.arange(7))
ax.set_yticklabels(axis_labels, fontsize=10)

# Eliminating the spine
ax.spines['right'].set_visible(False) # eliminating right border
ax.spines['top'].set_visible(False) # eliminating top border

# Loop to annotate values
for i in range(0, df_corr.shape[0]):
    for j in range(0, df_corr.shape[1]):
        plt.text(x=i, y=j, s=round(df_corr[i, j], 2), ha="center", va="center")

# Displaying Color Bar
cbar = ax.figure.colorbar(im, ax=ax)
cbar.ax.set_ylabel("Correlation", rotation=90, va="center")

plt.show()
```
