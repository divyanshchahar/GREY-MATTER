<h6> Importing Libraries </h6>

```python
import pandas as pd
import matplotlib.pyplot as plt
```

<h6> Creating Dataset</h6>

```python
df = pd.read_csv("world_population_data.csv", skiprows=4) #Reading Dataset

df.drop(["Country Code", "Indicator Name", "Indicator Code", "2020", "Unnamed: 65"], axis=1, inplace=True) # Dropping unnecessary columns

df = df.loc[df["Country Name"] == "India"] # Selecting Data of India

df.drop(["Country Name"], axis=1, inplace=True)

Y = df.values.tolist()[0] # Creating Values for Y-axis

#Creating Values for Y-axis
temp = df.columns.tolist()
X =[]
for member in temp:
    X.append(int(member))
```

<h6> Plotting Graph </h6>

```python
fig, ax = plt.subplots()

plt.plot(X, Y)

plt.title("Population Growth", fontsize=20, y=-0.25)

# X-axis Operations
plt.xticks(fontsize=10, rotation=0, ha='center') # tick-marks
ax.set_xlabel("Years", fontsize=12) # x-label

# Y-axis operations
plt.yticks(fontsize=10) # y-tick marks
ax.set_ylabel("Population", fontsize=12) #y-label

# Eliminating the spine
ax.spines['right'].set_visible(False) # eliminating right border
ax.spines['top'].set_visible(False) # eliminating top border

plt.show()
```
