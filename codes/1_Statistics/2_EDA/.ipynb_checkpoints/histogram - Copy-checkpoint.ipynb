<h2> <u> Univariate Graphical EDA </u> </h2>
<h3> <u> Bar Graph </u> </h3>

<h6> Importing Libraries </h6>

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

<h6> Creating Dataset</h6>

```python
sales = [100, 120, 90, 75, 60, 100, 130, 135, 150, 165, 190, 200]
month = ["January", "Feburary", "March", "April", "May", "June", "July", "August", "September", "OCtober", "November", "December"]
```

<h6> Plotting the Graph </h6>

```python
ofst = 0.5  # offset
xpos = np.arange(len(sales))  # x-positions

fig, ax = plt.subplots()

grp1 = ax.bar(xpos, sales)

plt.title("Car Sales for the year 2019-2020", fontsize=20)

# X-axis Operations
plt.xticks(fontsize=10, rotation=0, ha='center') # tick-marks
ax.set_xlabel("Month", fontsize=12) # x-label
ax.set_xticks(xpos) # x-tick location
ax.set_xticklabels(month, rotation=30, fontsize=10) # x-ticklabels

# Y-axis operations
plt.yticks(fontsize=10) # y-tick marks
ax.set_ylabel("Sales in Millions", fontsize=12) #y-label

# Eliminating the spine
ax.spines['right'].set_visible(False) # eliminating right border
ax.spines['top'].set_visible(False) # eliminating top border

plt.show()
```
