<h2> <u> Univariate Graphical EDA </u> </h2>

<h6> Importing Libraries </h6>

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
<h6> Reading Data</h6>
```

```python
df = pd.read_csv("ghg_by_sector.csv")
sectors = df["Sector"].values.tolist()
share = df["Share of global greenhouse gas emissions (%)"].values.tolist()
```

```python
plt.pie(share, # list of values to be plotted
        labels=sectors, #Labels
        autopct='%1.1f%%', # Marking the portion of pie with its share value
        explode=(0.1, 0, 0, 0), # Drawing out the biggest portin of the pi
        startangle=0, # Starting position for plotting the pie chart
        rotatelabels=True, # Rotating labels
        radius=2) # Size of the Graph

plt.title("Carbon Footprint of Various Industries", va="bottom")

plt.show()
```
