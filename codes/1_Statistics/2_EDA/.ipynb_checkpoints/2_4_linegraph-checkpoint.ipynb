<h2> <u> Line Graphs </u> </h2>

In this demo we will perform the following operations:

- Title:
    - set the title to required title and fontsize to 20

- x-axis operations:
    - set the x-tick marks fontsize to 10, rotation to 0 and horizontal alingment to center
    - set x-axis label to "Month" and fontszie to 12

- y-axis operations:
    - set the y-tick marks fontsize to 10
    - set y-label to required parameter and fontsize to 12
   
- Eliminate the top and right hand spine

- Legend Operations (only for Graph 2 and 3):
    - Set the legend as name of the countries, anchor the legend box horizontally in the center and 0.25 units below the graph and set the anchor point in the center

- Date operations (only for Graph-3):
    - Set the x-axis ticks as per dates

<h6> Importing Libraries </h6>

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import dates as mpl_dates
```

<h3> <u> 1. Line Graph (Univariate EDA) </u></h3>
<h6> Creating Dataset </h6>

```python
df = pd.read_csv("world_population_data.csv", skiprows=4) #Reading Dataset
df.drop(["Country Code", "Indicator Name", "Indicator Code", "2020", "Unnamed: 65"], axis=1, inplace=True) # Dropping unnecessary columns
```

<h6> Creating Variables </h6>

```python
temp = df.loc[df["Country Name"] == "India"] # Selecting Data of India
temp.drop(["Country Name"], axis=1, inplace=True) # Removing un-necessary data

X_1 =[]

# Loop to convert strings to integers
for member in temp.columns.tolist():
    X_1.append(int((member)))
    
Y_1 = temp.values.tolist()[0]
```

<h6> Plotting the Graph </h6>

```python
fig, ax = plt.subplots()

plt.plot(X_1, Y_1)

plt.title("Population Growth", fontsize=20)

# X-axis Operations
plt.xticks(fontsize=10, rotation=0, ha='center') # tick-marks
ax.set_xlabel("Years", fontsize=12) # x-label

# Y-axis operations
plt.yticks(fontsize=10) # y-tick marks
ax.set_ylabel("Population", fontsize=12) #y-label

# Eliminating the spine
ax.spines['right'].set_visible(False) # eliminating right border
ax.spines['top'].set_visible(False) # eliminating top border

plt.show()
```

<h3> <u> 2. Line Graph (Multivariate EDA) </u></h3>
<h6> Creating Variables </h6>

```python
temp = df.loc[df["Country Name"] == "United States"] # Selecting Data of India
temp.drop(["Country Name"], axis=1, inplace=True) # Removing un-necessary data

Y_2 = temp.values.tolist()[0]

Y = []

Y.append(Y_1)
Y.append(Y_2)

countries = ["India", "USA"]
```

<h6> Plotting the Graph </h6>

```python
run_count = len(Y)

fig, ax = plt.subplots()

for i in range(0, run_count):
    ax.plot(X_1, Y[i], label=countries[i])

plt.title("Population Growth", fontsize=20) # Title

plt.legend(fontsize=10, loc='center',ncol=2, bbox_to_anchor=(0.5, -0.3))  # Legend

# X-axis Operations
plt.xticks(fontsize=10, ha='center') # tick-marks
ax.set_xlabel("Years", fontsize=12) # x-label

# Dealing with dates
date_format = mpl_dates.DateFormatter('d%m%Y')
plt.gcf().autofmt_xdate()

# Y-axis operations
plt.yticks(fontsize=10) # y-tick marks
ax.set_ylabel("Population", fontsize=12) #y-label

# Eliminating the spine
ax.spines['right'].set_visible(False) # eliminating right border
ax.spines['top'].set_visible(False) # eliminating top border

plt.show()
```

<h3> <u> 3.Run Sequence Plot </u></h3>
<h6> Reading Dataset </h6>

```python
df_rs = pd.read_csv("GOOGL.csv")
df_rs.dropna(inplace=True)
```

<h6> Crearing Variables </h6>

```python
X_3 = pd.to_datetime(df_rs["Date"], dayfirst="True")
Y_3 = df_rs["High"]
```

<h6> Plotting the Graph </h6>

```python
fig, ax = plt.subplots()

ax.plot(X_3, Y_3)

plt.title("Google Stock Price", fontsize=20) # Title

# X-axis Operations
plt.xticks(fontsize=10, ha='center') # tick-marks
ax.set_xlabel("Date", fontsize=12) # x-label

# Dealing with dates
date_format = mpl_dates.DateFormatter('d%m%Y')
plt.gcf().autofmt_xdate()

# Y-axis operations
plt.yticks(fontsize=10) # y-tick marks
ax.set_ylabel("Opening Price", fontsize=12) #y-label

# Eliminating the spine
ax.spines['right'].set_visible(False) # eliminating right border
ax.spines['top'].set_visible(False) # eliminating top border

plt.show()
```
